```python
import geopandas
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import numpy
```

## Import and format of precinct data.
Unfortunately precincts vary greatly and not all are in the form of lat and lon. I will com back to this later.

```python
def import_and_format(path:str)->geopandas.GeoDataFrame():
    '''This function accepts the path to a folder containing
    a shapefilethen cleans and gives it a uniform formatting.
    It returns the geopandas dataframe with proper formatting.'''
    
    # Add slash to end of path if necessary
    if((path[-1]!='/') & (path[-1]!='\\')):
        path += '/'
    
    # Find shapefile
    file = [f.name for f in os.scandir(path) if f.name[-4:]=='.shp']
    
    assert(len(file)<2), 'Found multiple shapefiles!'
    assert(len(file)>0), 'Found no shapefiles!'
    
    # Load shapefile
    tdf = geopandas.read_file(path+file[0])
    tdf.columns = [c.lower() for c in tdf.columns]
    
    
    # Determine total number of votes
    vote_cols = [c for c in tdf.columns if c[:3]=='g16']
    tdf['total_votes'] = tdf[vote_cols].sum(axis=1)
    tdf['cli_percent'] = tdf['g16predcli'] / tdf['total_votes']
    tdf['tru_percent'] = tdf['g16prertru'] / tdf['total_votes']
    
    #Load name
    name_col=[c for c in tdf.columns if 'name' in c]
    if len(name_col) > 0:
        tdf['calc_name'] = tdf[name_col[0]]
    else:
        tdf['calc_name'] = 'NA'
    
    #save source
    tdf['source'] = 'file'
    return tdf

voi = ['total_votes','cli_percent','tru_percent','g16predcli','g16prertru','source','geometry']


root_file = 'source data/precinct data/'
df = geopandas.GeoDataFrame(columns=voi)
for folder in os.scandir(root_file):
    if(re.match(r'^[a-z]{2}_\d{4}$',folder.name)):
        target = root_file+'{name:}/{name:}.shp'.format(name=folder.name)
        
        print(folder.name)
        tdf = import_and_format(root_file+folder.name)[voi]
        tdf.plot()
        plt.show()
        
        df = df.append(import_and_format(root_file+folder.name)[voi])
        
df.plot()
```

## Importing county level data

```python
source = 'source data/us county shapefile/tl_2019_us_county.shp'

county = geopandas.read_file(source)
county['COUNTYFP'] = county['COUNTYFP'].astype(int)
county['GEOID'] = county['GEOID'].astype(int)

print(county.head())
county.plot()
```

The following block of code looks for data that cannot be matched from shapefiles to voter results.

```python
source = 'source data/county presidential results/countypres_2000-2016.csv'
county_vote = pd.read_csv(source).dropna(subset=['FIPS'])
county_vote['FIPS'] = county_vote['FIPS'].astype(int)
cv_voi = ['party','candidatevotes','totalvotes','FIPS']

mask = county_vote['year'] == 2016
c2016 = county_vote[mask].fillna(
    {'party':'na','candidatevotes':0}).pivot_table(values='candidatevotes',columns='party',index='FIPS')

vote_cols = c2016.columns
c2016 = c2016.reset_index().rename(columns={'FIPS':'GEOID'})
c2016
```

```python
cdf2016=county[['GEOID','geometry']].merge(c2016,on='GEOID')

cdf2016
```

This map is a logical check to see who won.

```python
mask = (cdf2016['republican'] > cdf2016['democrat']) & (cdf2016['republican'] > cdf2016['na'])
cdf2016.loc[mask,'color'] = 'Trump'

mask = (cdf2016['democrat'] > cdf2016['republican']) & (cdf2016['democrat'] > cdf2016['na'])
cdf2016.loc[mask,'color'] = 'Clinton'

plt.figure()

f,ax = plt.subplots(1,1)

f.set_figheight(15)
f.set_figwidth(15)

ax.set_xlim(-135,-60)
cdf2016.plot(column='color',ax=ax,
             legend=True,cmap='RdBu',
             categories=['Trump','Clinton']
            )
```
